A base foi retirada do <a href="https://www.kaggle.com/kaggle/kaggle-survey-2017">Kaggle Data Science Survey</a> que inclui respostas de mais de 10k pessoas que escrevem códigos e analizam dados na sua rotina diária de trabalho. </p>

```r
# Setando o working directory
dir("home/nbuser/library/kaggle-survey")
```

```r
# Carregando as libraries
library(tidyverse)               # Tidyverse é uma coleção que possui os pacotes "mais utilizados" em data science

# Carregando o dataset
responses <- read_csv("dataset/kagglesurvey.csv")

# Visualizando a base
head(responses,10)
```

## 2. Quais são as ferramentas?
<p>O objetivo é focar em quais ferramentas e linguagens as pessoas que responderam a pesquisa utilizam no seu trabalho. </p>

```r
# Novo dataframe
tools <- responses

# Splitar a feature "WorkToolsSelect" unnests the new column
tools <- tools  %>% 
    mutate(work_tools = strsplit(as.character(tools$WorkToolsSelect),",")) %>%
                                                                    unnest(work_tools)
# OBS.: A função strsplit() retorna uma lista com as palavras splitadas por qualquer "sep". A função unnest() pega a feature
# resultante e devolve o resultado informação criada pelo split.


head(tools)
```

## 3. Número de usuários por ferramenta
<p>Agora que todas as ferramentas foram splitadas para cada resposta, dá para ver quais ferramentas são as mais populares.</p>

```r
# Novo dataframe
tool_count <- tools

# Agrupando por "work_tools" e calculando o número de respostas para cada ferramenta
tool_count <- tool_count  %>% 
    group_by(work_tools)  %>% 
    summarise(numberOfResponses = n())

# Ordenando
tool_count <- tool_count[order(tool_count$numberOfResponses,decreasing = T),]


head(tool_count)
sum(tool_count$numberOfResponses)
```

## 4. Plotando as ferramentas mais populares

```r
#Setando as dimensões do gráfico
options(repr.plot.width = 10, repr.plot.height=8)

# Gráfico de barras com a coluna "work tools".
ggplot(tool_count, aes(x=reorder(work_tools,-numberOfResponses), 
                       y=numberOfResponses)) + 
    geom_bar(stat="identity",fill = "#0399D3") +
    theme(axis.text.x = element_text(angle = 90, hjust = 1)) + 
    ggtitle("Distribuição das Ferramentas Preferidas em Ciência de Dados") + xlab("Ferramentas") + 
    ylab("Número de Respostas") + 
    geom_text(aes(label=numberOfResponses), hjust = 0.5, vjust = -0.5, size = 2, color = 'black') + 
    theme(plot.title = element_text(lineheight=.9, face="bold",hjust = 0.5))
    
```

Como vemos, as três mais utilizadas são Python, R e SQL. De todas as ferramentas que os cientistas de dados disseram que mais usam, 12% deles disseram que mais usam o Python, 10% mais usam o R e 9% mais usam o SQL. Python tem 29% a mais de usuários do que o R e 42% a mais do que o SQL.

## 5. R vs Python
<p>Dentro da ciência de dados, existe um debate entre os profissionais sobre qual a melhor linguagem, R ou Python. O objetivo é fazem comparações das impressões dos usuários em ambas.</p>

```r
# Novo dataframe
debate_tools <- responses

# Criando coluna de preferência "language_preference"
debate_tools <- debate_tools  %>% 
   mutate(language_preference = case_when(
        grepl("R",WorkToolsSelect) & !grepl("Python",WorkToolsSelect) ~ "R",
        grepl("Python",WorkToolsSelect) & !grepl("R",WorkToolsSelect) ~ "Python",
        grepl("Python",WorkToolsSelect) & grepl("R",WorkToolsSelect) ~ "Both",
        !grepl("Python",WorkToolsSelect) & !grepl("R",WorkToolsSelect) ~ "Neither",))

# OBS.: case_when() faz parte da biblioteca dplyr e é um "if else mais inteligente"

head(debate_tools)
```

## 6. Plotando R vs Python

```r
# Setando as dimensões do gráfico
options(repr.plot.width = 10, repr.plot.height=3.5)

# Novo dataframe
debate_plot <- debate_tools

# Agrupando por preferência
debate_plot <- debate_plot  %>% 
    group_by(language_preference)  %>% 
    summarise(numberOfResponses = n())  %>% 

# Removendo os usuários que responderam "neither" (nenhum)
    filter(language_preference!="Neither")

# Plotando o gráfico de barras
    ggplot(debate_plot,aes(x=reorder(language_preference,-numberOfResponses)
                       ,y=numberOfResponses)) + 
    geom_bar(stat="identity",fill = "#0399D3") + 
    ggtitle("Distribuição: R vs Python") + xlab("Ferramenta") + 
    ylab("Número de Respostas") + 
    geom_text(aes(label=numberOfResponses), hjust = 0.5, vjust = 1.5, size = 5, color = 'black') + 
    theme(plot.title = element_text(lineheight=.9, face="bold",hjust = 0.5))
```

Do total de pessoas que disseram que usam as linguagens, 50% das pessoas usam as duas linguagens enquanto temos 33% para só Python e 17% para R.

## 7. Recomendações
<p>A maioria dos profissionais programam tanto em R quanto em Python. Mas o que acontece se for perguntado para eles quais linguagens eles recomendariam para os newbies? </p>

```r
debate_tools
```

```r
# Novo dataframe
recommendations <- debate_tools

# Agrupando por preferência e por recomendação
recommendations <- recommendations  %>% 
    group_by(language_preference,LanguageRecommendationSelect)  %>% 
    summarise(numberOfResponses=n()) %>%

# Removendo respostas em branco e incluindo as top recomendações
                    filter(LanguageRecommendationSelect!="") %>%
                    arrange(language_preference,desc(numberOfResponses)) %>%
                    mutate(counts = row_number()) %>%
                    filter(counts<=4)
                    
```

## 8. Linguagem mais recomendada de acordo com a preferência
<p>Plotando quais linguagens são as mais recomendadas baseadas nas linguagens que uma pessoa utiliza.</p>

```r
recommendations
```

```r
# Setando as dimensões do gráfico
options(repr.plot.width = 10, repr.plot.height=7)

# Gráfico de barras segmentado
ggplot(recommendations, aes(x=reorder(LanguageRecommendationSelect,-numberOfResponses)
                            ,y=numberOfResponses)) + 
            geom_bar(stat="identity", fill = "#0399D3") + facet_wrap(~language_preference) + 
            theme(axis.text.x = element_text(angle = 90, hjust = 1)) + 
            ggtitle("Quais linguagens São Recomendadas de Acordo com a Preferência") + 
            xlab("Linguagem Recomendada") + 
            ylab("Número de Respostas") + 
            geom_text(aes(label=numberOfResponses), hjust = 0.5, vjust = -0.1, size = 2.4, color = 'black') + 
            theme(plot.title = element_text(lineheight=.9, face="bold",hjust = 0.5))
```
